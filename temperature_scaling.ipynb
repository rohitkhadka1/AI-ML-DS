```python
import numpy as np
import matplotlib.pyplot as plt

# Base logits
logits = np.array([2.0, 1.0, 0.1])

def softmax_temp(z, T):
    z_stable = z / T
    z_stable = z_stable - np.max(z_stable)
    exp_z = np.exp(z_stable)
    return exp_z / np.sum(exp_z)
```

```python
# Temperature values to test
temps = [0.5, 1.0, 2.0, 5.0]
all_probs = [softmax_temp(logits, T) for T in temps]

# Plot: Softmax at different temperatures
plt.figure()
for i, T in enumerate(temps):
    plt.bar(np.arange(3) + i*0.2, all_probs[i], width=0.2, label=f"T={T}")

plt.title("Softmax Distribution at Different Temperatures")
plt.legend()
plt.show()

```

```python
# Extra: Plot logits / T directly to visualize flattening or sharpening
plt.figure()
for i, T in enumerate(temps):
    plt.plot(logits / T, marker='o', label=f"T={T}")

plt.title("Logits Divided by Temperature (Before Softmax)")
plt.legend()
plt.show()
```
